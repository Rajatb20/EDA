```python
df = data.drop(['ConfirmedForeignNational','ConfirmedIndianNational'], inplace = True, axis = 1)
```

```python
data.head()
```

```python
data['ActiveCases'] = data['Confirmed'] - data['Cured'] - data['Deaths']
```

```python
data.tail(10)
```

```python
# top 10 states with highest active cases
df1 = data.groupby(by = 'State/UnionTerritory').max()[['ActiveCases','Date']].sort_values(by = ['ActiveCases'], ascending = False).reset_index()
```

```python
plt.figure(figsize=(15,10))
plt.title(' top 10 states with highest active cases', size = 20)
sns.barplot(data = df1.iloc[:10],y="ActiveCases",x="State/UnionTerritory",edgecolor = 'black')
plt.xlabel('State/UnionTerritory')
plt.ylabel('ActiveCases')
plt.show()
```

```python
# top 5 states with highest deaths
df2 = data.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
```

```python
plt.figure(figsize=(10,7))
plt.title(' top 10 states with highest deaths', size = 20)
sns.barplot(data = df2.iloc[:7], y="Deaths", x="State/UnionTerritory", edgecolor = 'red')
plt.xlabel('State/UnionTerritory')
plt.ylabel('Deaths')
plt.show()
```

```python
# Growth trend
plt.figure(figsize=(7,10))
sns.lineplot(data = data[data['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Tamil Nadu','Kerala','Uttar Pradesh']),x ="Date",y ="ActiveCases",hue ="State/UnionTerritory"])
plt.title('GROWTH TREND',size=15)
```
